```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#导入数据，绘制散点图查看数据
data = np.genfromtxt("C:/Users/Yuang/Desktop/机器学习数据/data.csv",delimiter=",")
x_data = data[:,0]
y_data = data[:,1]
plt.scatter(x_data,y_data)
plt.show()
```

```python
#学习率
lr = 0.0001
#截距
b = 0
#斜率
k = 0
#最大迭代次数
epochs = 50

#最小二乘法计算代价函数
def compute_error(b,k,x_data,y_data):
    total_error = 0
    for i in range(0,len(x_data)):
        total_error += ((b+(k*x_data[i]))-y_data[i])**2
    return total_error/float(len(x_data))/2.0

#梯度下降
def gradient_decent_runner(x_data,y_data,lr,b,k,epochs):
    m = float(len(x_data))
    
    #循环epochs次
    for i in range(epochs):
        b_grad = 0
        k_grad = 0
        
        for j in range(0,len(x_data)):
            b_grad += (1/m)*((b+(k*x_data[j]))-y_data[j])
            k_grad += (1/m)*x_data[j]*((b+(k*x_data[j]))-y_data[j])
        #更新b,k
        b = b-(lr*b_grad)
        k = k-(lr*k_grad)
        
        #每迭代五次输出一次图像
        if i % 5==0:
            print("epoch = ",i)
            plt.plot(x_data,y_data,'b.')
            plt.plot(x_data,k*x_data+b,'r')
            plt.show()
    return b,k      
```

```python
print('starting b = {0}, k = {1},error = {2}'.format(b, k, compute_error(b,k,x_data,y_data)))
print('Running...')
b,k = gradient_decent_runner(x_data,y_data,lr,b,k,epochs)
print('after {0} interations, b ={1}, k ={2},error = {3}'.format(epochs, b, k, compute_error(b,k,x_data,y_data)))
```
