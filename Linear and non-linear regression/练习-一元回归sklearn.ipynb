```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("C:/Users/Yuang/Desktop/机器学习数据/data.csv",delimiter = ",")
x_data = data[:,0]
y_data = data[:,1]
plt.scatter(x_data,y_data)
plt.show()
print(x_data.shape)
```

```python
#按照sklearn库的要求处理数据
x_data = data[:,0,np.newaxis]
y_data = data[:,1]

#创建并拟合模型
model = LinearRegression()
model.fit(x_data,y_data)
```

```python
#画图
plt.plot(x_data,y_data,"r.")
plt.plot(x_data,model.predict(x_data),"y")
plt.show()
```
